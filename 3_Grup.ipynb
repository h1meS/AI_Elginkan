```python
import pandas as pd
import numpy as np
data=pd.read_csv("Job_Placement_Data.csv")
type(data)
print(data)
```

```python
data.info()
```

```python
print(data.corr())
```

```python
XSample=data.iloc[:,:-1] 
ySample=data.status
print(XSample)
print(ySample)
```

# LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
XSample.gender=le.fit_transform(XSample.gender)
XSample.ssc_board=le.fit_transform(XSample.ssc_board)
XSample.hsc_board=le.fit_transform(XSample.hsc_board)
XSample.hsc_subject=le.fit_transform(XSample.hsc_subject)
XSample.undergrad_degree=le.fit_transform(XSample.undergrad_degree)
XSample.work_experience=le.fit_transform(XSample.work_experience)
XSample.specialisation=le.fit_transform(XSample.specialisation)
ySample=le.fit_transform(ySample)
print(ySample)

```

```python
import matplotlib.pyplot as plt
from random import randint as rnd

plt.subplot(3,3,1)
plt.plot(XSample.gender)

plt.subplot(3,3,3)
plt.plot(XSample.work_experience)

plt.subplot(3,3,7)
plt.plot(XSample.ssc_board)
plt.show()
```

```python
XSample.corr()
```

```python
from sklearn.model_selection import train_test_split
XTrain, XTest, yTrain, yTest = train_test_split(XSample,ySample, test_size=0.20 , random_state=21)
# neye ayırıcaz train, test = ayırma komutu( ayırılacak değişkenler , test size ile kaç kaç ayırıcazz yüzdelkik)
print(XTrain,XTest,yTrain,yTest)
print(XTrain.shape,XTest.shape,yTrain.shape,yTest.shape)
```

#### ÖLÇEKLENDİRMEYE DİKKAT STANDARDSCALER VEYA MİN_MAX_SCALER 

```python
#STANDARDSCALER
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
XTrain=sc.fit_transform(XTrain)
XTest=sc.fit_transform(XTest)
print("XTrain")
print(XTrain)
print("XTest")
print(XTest)
```

```python
#min_max_scaler 
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
XTrain=mms.fit_transform(XTrain)
XTest=mms.fit_transform(XTest)
print(XTest.shape)
print(XTrain.shape)
print(XTrain)
print(XTest.max(),XTest.min(), yTest.max(), yTrain.min())
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=200, max_depth=7, random_state=21) #n_estimate ile kaç tane karar ağacını karşılaştıracağını elirler, max_depth derinliği verir
classifier.fit(XTrain, yTrain)
yTestPred=classifier.predict(XTest)
print(yTest)
print("ytestPred")
print(yTestPred)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import  classification_report, ConfusionMatrixDisplay, confusion_matrix 
yTestPred=classifier.predict(XTest)
print(classification_report(yTest,yTestPred))
cm=confusion_matrix(yTest, yTestPred, normalize="true").round(2)
disp=ConfusionMatrixDisplay(confusion_matrix=cm)
disp=disp.plot(cmap=plt.cm.GnBu, values_format='g')
plt.title(f"Train confusion matrix",style="italic")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=1.5)  # "p= minkowski katsayısı" p=1 manathan, p=2 öklid, p=sonsuz chebyshev
classifier.fit(XTrain,yTrain) # eğitim fonksiyonuna ayırma
yTrainPred=classifier.predict(XTrain)
print(yTrainPred) #yTrain tahmin değerleri
print(yTrain) #yTrain gerçek değerlerimiz
```

```python
#modeldeki eğitimi test etme
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix
print(classification_report(yTrain, yTrainPred))
cm=confusion_matrix(yTrain, yTrainPred, normalize="true").round (2)
#cmn=cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] 
#cmn=cmn.round(2)
disp=ConfusionMatrixDisplay(confusion_matrix=cm) 
disp=disp.plot(cmap=plt.cm.GnBu, values_format='g')
plt.title(f"Train Confusion Matrix", style="italic") 
#plt.savefig(f "Model2_CM_8901_(self.header)_Data_Son.jpg")
plt.show()
```

```python
from sklearn.svm import SVC
classifier=SVC(C=1.6, kernel="linear", gamma=0.1) #cfloat, default=1.0 kernel{"linear","poly","rbf","sigmoid","precomputed"}
#or callable, default="rbf", gamma{"scale","auto"} or float, default="scale"
classifier.fit(XTrain, yTrain)
yTestPred=classifier.predict(XTest)
print(yTest)
print("ytestPred")
print(yTestPred)

print(classification_report(yTest,yTestPred))
cm=confusion_matrix(yTest, yTestPred, normalize="true").round(2)
disp=ConfusionMatrixDisplay(confusion_matrix=cm)
disp=disp.plot(cmap=plt.cm.GnBu, values_format='g')
plt.title(f"Train confusion matrix",style="italic")
plt.show()
```
