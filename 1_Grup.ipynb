```python
def find_unique_values(col_name):
    result = []

    for value in dataset[col_name]:
        if value not in result:
            result.append(value)
    
    return result
```

```python
class Labelizer:
    def __init__(self, cols) -> None:
        self.cols = cols
        self.current_col = ""

    def labelize(self, value):
        unique_val = find_unique_values(self.current_col)

        for i in range(len(unique_val)):
            if value == unique_val[i]:
                return i
    
    def labelize_all(self):
        for col in self.cols:
            self.current_col = col
            dataset[col] = dataset[col].apply(self.labelize)
```

```python
labelizer = Labelizer(["ssc_percentage", "ssc_board", "hsc_board", "undergrad_degree", "work_experience", "specialisation", "hsc_subject"])
labelizer.labelize_all()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

XSample = sc.fit_transform(XSample)
```

```python
from sklearn.model_selection import train_test_split

XTrain, XTest, yTrain, yTest = train_test_split(XSample, ySample, test_size=0.10, random_state=42)
```

```python
XTrain
```

```python
from  sklearn.neighbors import  KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)#paantez içi metrik 2 öklid 1 menethen sonsuz spi
classifier.fit(XTrain,yTrain)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100, max_depth = 7, random_state=9)
classifier.fit(XTrain, yTrain)
yTestPred=classifier.predict(XTest)
print(yTest) 
print(yTestPred)
```

```python
from sklearn.svm import SVC #clasification olduğu için svc
classifier=SVC(C=2, kernel="sigmoid", gamma=0.3)
#Cfloat, default=1.0, kernel{'linear', 'poly', 'sigmoid','precomputed'} 
#or callable, default='rbf', gama{'scale','auto}
classifier.fit(XTrain, yTrain)
yTestPred=classifier.predict(XTest) 
print(yTestPred)
print(yTest)
#mavi ekran bunun altında olacak
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix

yTestPred=classifier.predict(XTest)
print(classification_report(yTest, yTestPred))
cm=confusion_matrix(yTest, yTestPred, normalize="true").round(2) 
#cmn = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
#cmn=cmn.round (2)
disp=ConfusionMatrixDisplay(confusion_matrix=cm)
disp = disp.plot(cmap=plt.cm.GnBu, values_format='g') 
plt.title(f"Train Confusion Matrix", style="italic") 
#plt.savefig(f"Mode12_CM_8901_{self.header} _Data_Son.jpg") 
plt.show()
```
